# Основные выводы и рекомендации
Основная проблема - сложность в ориентировании на сайте
Способы решения:
1. Добавление фильтров
2. Сегментация главной страницы по блокам
3. Добавление карты   

### Основные идеи фич: 
фильтры по цене, наличию билетов
добавить карту (общую, либо под каждым мероприятием)

## 1 пользователь: студент, 18-23 лет, использует Культура:
**1. Как часто вы покупаете билеты на события (концерты, театры, спортивные мероприятия и т.д.)?**  
-- 1 раз в месяц  

**2. Как вы обычно узнаете о предстоящих событиях?**  
-- при помощи приложения: культура  

**3. Какие факторы влияют на ваше решение о покупке билета?**  
-- цена, содержание мероприятия, время мероприятия  

**4. Почему вы выбираете именно приложение "Культура"?**  
-- визуальная часть, карта, разделение мероприятий по категориям на главной странице (мастер-классы, квесты)  

**5. С какими трудностями вы сталкивались при поиске мероприятия? Как вам удалось их решить?**   
-- мало представлено интересных мероприятий, решала поискав в интеренете другие мероприятия  

**6. Есть ли что-то, что вы хотели бы изменить в текущем сервисе или приложении?**    
-- функции, которые бы хотела добавить: видео с мероприятия, анимации на сайте, чтобы под каждым мероприятием можно было посмотреть, где находится это мероприятие на карте  

**7. Что вам не нравится/раздражает при поиске билетов?**  
-- слишком много визуального шума на некоторых сайтах  

**8. В каком разделе вы чаще находите интересное для вас мероприятие? (популярное, ближайшие, либо ищете по конкретным категориям: театры, кино и т.п.)**  
-- По категориям: кино, спектакли

## 2 пользователь: работник, 25-30 лет, использует Культура, VL.ru:
**1. Как часто вы покупаете билеты на события (концерты, театры, спортивные мероприятия и т.д.)?**  
-- 1-2 раза в месяц, пока была Пушкинская покупал 2 раза в месяц  

**2. Как вы обычно узнаете о предстоящих событиях?** 
-- При помощи сайтов афиш: "Культура" в основном, "VL.ru"  

**3. Какие факторы влияют на ваше решение о покупке билета?**  
-- Содержание мероприятия, цена, не нужно переходить на сторонние сайты, можно с легкостью оплатить картой  

**4. Почему вы в основном выбираете именно приложение "Культура"?**  
-- Потому что это приложение, через браузер не всегда удобно  

**5. С какими трудностями вы сталкивались при поиске мероприятия? Как вам удалось их решить?**   
-- Не грузится сайт для оплаты, не принимается моя карта, цена оверпрайс. Решение - забить на мероприятие, купить билет офлайн  

**6. Есть ли что-то, что вы хотели бы изменить в приложении "Культура"?**    
-- Сделать фильтры: дата, место проведение, цена, наличие билетов. Сделать выбор региона в культуре более удобным (город, а не край). Сделать разделы с категориями в меню.  

**7. Есть ли что-то, что вы хотели бы изменить в сайте "VL.ru"?**   
-- Разделить события, рассортировать, изменение дизайна  

**8. В каком разделе вы чаще находите интересное для вас мероприятие? (популярное, ближайшие, либо ищете по конкретным категориям: театры, кино и т.п.)**  
-- По категориям: кино
